```python
import pandas as pd
import random
```

```python
verse_df = pd.read_pickle('data/gutenberg_rythm_ano.pkl')
```

```python
gutenberg_df = pd.read_csv('data/gutenberg.csv')
```

```python
gutenberg_df.head()
```

```python
verse_df['title'] = gutenberg_df['title']

verse_df_grouped = verse_df.groupby('title').agg(list).sample(frac=1).reset_index()
```

```python
verse_df_grouped.head()
```

```python
max_syll = 32
min_syll = 4
title_lst = []
verse_lst = []
syll_count_lst = []
strophe_lst = []
verse_idx_lst = []
len_poem_lst = []

verse_idx = 0
last_title = None
for index, row in verse_df_grouped.iterrows(): 

    if row['title'] in title_lst:
            title = row['title'] + str(index)
    else: 
        title = row['title']
    strophe_idx = 0
    verse_idx = 0
    poem_length = 0
    for strophe in row['text']:
        poem_length += len(strophe)

    for i, verse in enumerate(row['text']):
        if verse:
            for j, line in enumerate(verse):
                if len(row['rythm'][i][j]) <= max_syll and len(row['rythm'][i][j]) >= min_syll:
                    title_lst.append(title)
                    verse_lst.append(line)
                    syll_count_lst.append(len(row['rythm'][i][j]))
                    strophe_lst.append(strophe_idx)
                    verse_idx_lst.append(verse_idx)
                    len_poem_lst.append(poem_length)
                    verse_idx += 1   
            strophe_idx += 1
```

```python
verse_data_df = pd.DataFrame(list(zip(strophe_lst,verse_idx_lst,title_lst,verse_lst,syll_count_lst,len_poem_lst)),
                columns =['strophe_idx','verse_idx','title','verse','syll_count','poem_length'])

verse_data_df.head(25)
```

```python
def get_syll_count_df(verse_data_df):
    syll_count_df = pd.DataFrame(verse_data_df['syll_count'])
    syll_count_df['orig_index'] = syll_count_df.index
    return syll_count_df.groupby('syll_count').agg(list)
```

```python
frac_beginnings = 4
frac_same_poem = 5
```

```python
def get_pairs_start(verse_data_df,num_verses,frac_beginnings,frac_same_poem):   

    poem_beginnings = verse_data_df.where((verse_data_df.verse_idx == 0) & (verse_data_df.poem_length > num_verses)).dropna().index.values.tolist()

    beginnings_sample = random.sample(poem_beginnings,int(len(poem_beginnings)/frac_beginnings))
    random.shuffle(beginnings_sample)

    beginnings_sample_match = beginnings_sample[:int(len(beginnings_sample)/2)]
    beginnings_sample_no_match = beginnings_sample[int(len(beginnings_sample)/2):]

    match_lst = []
    no_match_lst = []
    differences = []
    for item in beginnings_sample_match:
        row_1 = verse_data_df.iloc[item+num_verses-1]                   # the syllable data is taken from the last sentence
        verse_1 = ''
        for i in range(num_verses):
            verse_1 += ' ' + verse_data_df.iloc[item+i]['verse']

        row_2 = verse_data_df.iloc[item+num_verses]
        match = [verse_1,row_2['verse']]
        differences.append(row_1['syll_count']-row_2['syll_count'])
        match_lst.append(match)

    partners = []
    for i in range(1,num_verses):
        partners += [x+i for x in beginnings_sample_match]

    verse_data_df_2 = verse_data_df.drop(index=partners+beginnings_sample_match+beginnings_sample_no_match).reset_index(drop=True)

    idx = 0
    cnt = 0
    beginnings_sample_no_match_2 = beginnings_sample_no_match.copy()
    while cnt < len(beginnings_sample_no_match)/frac_same_poem:
        index = beginnings_sample_no_match[idx]
        title = verse_data_df.iloc[index]['title']
        candidates = verse_data_df_2.where((verse_data_df_2['title'] == title) & (verse_data_df_2['verse_idx'] > 5+num_verses)).dropna().index.values.tolist()
        idx += 1

        if candidates:

            chosen = random.sample(candidates,1)[0]
            cnt += 1
            beginnings_sample_no_match_2.remove(index)
            verse_1 = ''
            for i in range(num_verses):
                verse_1 += ' ' + verse_data_df.iloc[index+i]['verse']
            verse_2 = verse_data_df_2.iloc[chosen]['verse']
            no_match_lst.append([verse_1,verse_2])
            verse_data_df_2 = verse_data_df_2.drop(index=chosen).reset_index(drop=True)

    cnt = 0
    failure_cnt = 0
    for index in beginnings_sample_no_match_2:
        title = verse_data_df.iloc[index]['title']
        target_syll = differences[cnt] + verse_data_df.iloc[index+num_verses-1]['syll_count']
        cnt += 1
        candidates = verse_data_df_2.where((verse_data_df_2['title'] != title) & (verse_data_df_2['syll_count'] == target_syll) & (verse_data_df_2['verse_idx'] > 3)).dropna().index.values.tolist()
        if candidates:
            chosen = random.sample(candidates,1)[0]
            verse_1 = ''
            for i in range(num_verses):
                verse_1 += ' ' + verse_data_df.iloc[index+i]['verse']
            verse_2 = verse_data_df_2.iloc[chosen]['verse']
            no_match_lst.append([verse_1,verse_2])
            verse_data_df_2 = verse_data_df_2.drop(index=chosen).reset_index(drop=True)
        else:
            failure_cnt += 1

    print('number of failed pairs: ' + str(failure_cnt))

    return verse_data_df_2, match_lst, no_match_lst

```

```python

verse_data_df_2, match_lst_first, no_match_lst_first = get_pairs_start(verse_data_df,1,frac_beginnings,frac_same_poem)
```

```python
verse_data_df_3, match_lst_second, no_match_lst_second = get_pairs_start(verse_data_df_2,2,frac_beginnings,frac_same_poem)
```

```python
def get_n_verses(verse_data_df,num_verses,frac_same_poem):
    idx = 0
    idxs_complete = []
    idxs_partial = [] 
    verse_idx_lst = verse_data_df['verse_idx'].values.tolist()
    while idx < len(verse_data_df)-num_verses-1:
        lst = verse_idx_lst[idx:idx+num_verses]
        if sorted(lst) == verse_idx_lst[idx:idx+num_verses]:
            lst = lst = verse_idx_lst[idx:idx+num_verses+1]
            if sorted(lst) == verse_idx_lst[idx:idx+num_verses+1]:
                idxs_complete.append(idx)
                idx += num_verses + 1
            else:
                idxs_partial.append(idx)     # more randomness; to avoid synchronisation with strophes
                idx += num_verses 
                
            if random.random() < 0.3:
                idx += 1

        idx += 1
    
    total_preceeding_count = len(idxs_complete) + len(idxs_partial)

    print(total_preceeding_count)

    matches_count = min(len(idxs_complete),total_preceeding_count/2)

    random.shuffle(idxs_complete)

    match_lst = []
    differences = []

    idxs_complete_2 = idxs_complete.copy()

    idxs_to_delete = []
    for idx in range(int(matches_count)):
        index = idxs_complete[idx]
        verse_1 = ''
        for j in range(num_verses+1):
            verse_1 += ' ' + verse_data_df.iloc[index+j]['verse']

        verse_2 = verse_data_df.iloc[index+num_verses+1]['verse']

        difference = verse_data_df.iloc[index+num_verses]['syll_count'] - verse_data_df.iloc[index+num_verses+1]['syll_count']
        differences.append(difference)
        match_lst.append([verse_1,verse_2])
        idxs_to_delete += list(range(idx,idx+num_verses+1))
        try:
            idxs_complete_2.remove(index)
        except: 
            pass

    idxs_left = idxs_complete_2 + idxs_partial
    for idx in idxs_left:
        idxs_to_delete += list(range(idx,idx+num_verses))

    verse_data_df_2 = verse_data_df.drop(index=idxs_to_delete).reset_index(drop=True)

    cnt = 0
    idx = 0

    idxs_left_2 = idxs_left.copy()

    no_match_lst = []
    while cnt < len(idxs_left)/frac_same_poem:
        index = idxs_left[idx]
        title = verse_data_df.iloc[index+num_verses]['title']
        verse_idx = verse_data_df.iloc[index+num_verses]['verse_idx']
        candidates = verse_data_df_2.where((verse_data_df_2['title'] == title) & (abs(verse_data_df_2['verse_idx'] - verse_idx)>num_verses+1)).dropna().index.values.tolist()

        if candidates:
            chosen = random.sample(candidates,1)[0]
            cnt += 1
            idxs_left_2.remove(index)
            verse_1 = ''
            for i in range(num_verses):
                verse_1 += ' ' + verse_data_df.iloc[index+i]['verse']
            verse_2 = verse_data_df_2.iloc[chosen]['verse']
            no_match_lst.append([verse_1,verse_2])
            verse_data_df_2 = verse_data_df_2.drop(index=chosen).reset_index(drop=True)
        idx += 1 


    cnt = 0
    failure_cnt = 0

    for index in idxs_left_2:
        title = verse_data_df.iloc[index]['title']
        target_syll = differences[cnt] + verse_data_df.iloc[index+num_verses]['syll_count']
        cnt += 1
        candidates = verse_data_df_2.where((verse_data_df_2['title'] != title) & (verse_data_df_2['syll_count'] == target_syll)).dropna().index.values.tolist()
        if candidates:
            chosen = random.sample(candidates,1)[0]
            verse_1 = ''
            for i in range(num_verses):
                verse_1 += ' ' + verse_data_df.iloc[index+i]['verse']
            verse_2 = verse_data_df_2.iloc[chosen]['verse']
            no_match_lst.append([verse_1,verse_2])
            verse_data_df_2 = verse_data_df_2.drop(index=chosen).reset_index(drop=True)
        else:
            failure_cnt += 1

    return match_lst, no_match_lst
```
